```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path_new = 'Simulation result for Figure 2k.xlsx'  # Replace with the path to your file
sheet_data_new = pd.read_excel(file_path_new, sheet_name='0-to-1-cm-from-tegaderm')

# Extract time, distance, and data values
time_new = sheet_data_new['Time (min)'].values  # Convert to numpy array
distance_new = sheet_data_new.columns[1:].astype(float).values  # Convert to numpy array
values_new = sheet_data_new.iloc[:, 1:].values  # Extract data values

# Create the correct grid for pcolormesh
X, Y = np.meshgrid(time_new, distance_new)

# Create the heatmap using pcolormesh
plt.figure(figsize=(14, 2), dpi=600)  # High resolution for publication
plt.pcolormesh(
    X, Y, values_new.T, 
    cmap='jet', shading='auto', vmin=0, vmax=0.8
)
plt.colorbar(label='Value')

# Setting x and y ticks
plt.xticks(np.arange(0, time_new.max()+1, 120))  # 60 min intervals for x-axis
plt.yticks(np.arange(distance_new.min(), distance_new.max() + 1, 2))  # 1 mm intervals for y-axis


# Remove gridlines
plt.grid(False)

# Set labels and title
plt.xlabel('Time (min)')
plt.ylabel('Distance (mm)')
plt.title('Heat Map of Data with Uneven X-Axis Spacing')

# Save the plot as a high-resolution image
plt.savefig('heat_map_corrected_high_res.png', bbox_inches='tight', dpi=600)
plt.show()



```
